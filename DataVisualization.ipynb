```python

with sns.hls_palette(8, l=.5, s=1.):
    locationPlot = sns.catplot(kind="count",data=df, x = "Location", col="Fitness Center", facet_kws=dict(sharey=False))
    locationPlot.set_xlabels('Location', fontsize=10)
    locationPlot.set_ylabels('Number of Apartments', fontsize=10)
    plt.xlabel("Location")
    plt.show()
    ax = sns.countplot(y="Location", hue="Fitness Center", data=df, palette="muted")
    plt.show()
```

```python
with sns.hls_palette(8, l=.7, s=1.):
    relativePlot = sns.relplot(data=df, x="Distance", y="Cost", col="Location", kind="scatter", facet_kws=dict(sharex=False), ci=False, col_wrap=3)
    relativePlot.set_xlabels("Distance from Texas A&M campus(mi)")
    relativePlot.set_ylabels("Rent price")
    #relativePlot.set_titles("Cost($) of Apartments with distance(mi)")
    relativePlot.axes[0].set_xlim(0.0,10.0)
    relativePlot.axes[1].set_xlim(0.0,4.0)
    relativePlot.axes[2].set_xlim(0.0,3.0)
    relativePlot.axes[3].set_xlim(0.0,1.0)
    relativePlot.axes[4].set_xlim(0.0,3.0)
    plt.tight_layout()
    plt.show()
```
