```python
#Handles cost issues!
with open("dataset.csv", 'r') as f, open('datasetEdited.csv','w') as writer:
    sheet_reader = csv.reader(f, delimiter=",")
    sheet_writer = csv.writer(writer, delimiter=",")
    sheet_writer.writerow(["ApartmentName","URL","Beds","Baths","Cost","SquareFeet",
                        "Library","Fitness Center","Distance","Location"])
    next(sheet_reader)
    areaDict = {}
    for line in sheet_reader:
        listobj = []
        listobj.append(line[0]) #AptName
        listobj.append(line[1]) #URL
        
        #change studio to 1 bedroom
        if line[2] == "Studio":
            line[2] = "1"
            listobj.append(1)
        else:
            listobj.append(line[2])
            
        listobj.append(line[3]) #Baths
                
        #Handle total cost and remove $ sign
        #line[4] is string, chars is a list
        chars = [letter for letter in line[4]]
        if " " in chars:
            while " " in chars:
                chars.remove(" ")
        if "$" in chars:
            while '$' in chars: #double dollar signs
                chars.remove('$')
        if "," in chars:
            while "," in chars:
                chars.remove(",")
        if '-' in chars: #Range of vals
            if 'Bedroom' in line[4]: #per bedroom cost with range
                hyphen_position = chars.index("-")
                slash_position = chars.index("/")
                leftVal = ''.join(chars[:hyphen_position])
                rightVal = ''.join(chars[hyphen_position+1:slash_position])
                mean = math.ceil((int(leftVal) + int(rightVal)) / 2)
                total_cost = mean * int(line[2]) #Need to remove studio apartments though!\
                listobj.append(total_cost)
            else: #total cost with range 
                position = chars.index("-") #position of -
                leftVal = ''.join(chars[:position])
                rightVal = ''.join(chars[position+1:])
                #Both are strings, we calculate mean and round up resulting float
                mean = math.ceil((int(leftVal) + int(rightVal)) / 2)
                listobj.append(mean)
        else: #no range of vals
            if 'Bedroom' in line[4]: #per bedroom cost without range
                slash_position = chars.index("/")
                val = ''.join(chars[:slash_position])
                if line[2] == "Studio":
                    total_cost = int(val)
                    listobj.append(int(total_cost)) # no float conversion needed
                else:
                    total_cost = int(val) * int(line[2])
                    listobj.append(int(total_cost)) # no float conversion needed
            else:
                val = ''.join(chars)
                listobj.append(int(val)) # no float conversion needed
                
        #Handle range of values for SquareFeet
        lineFive = line[5]
        lineFive = lineFive.replace(',','')
        if lineFive != "":
            area = 0
            if '-' in lineFive:
                position = lineFive.index("-")
                val1 = int(lineFive[:position])
                val2 = int(lineFive[position+1:])
                area = math.ceil((val1 + val2) / 2)
                listobj.append(area)
            else:
                area = int(lineFive)
                listobj.append(lineFive)
            #write information for areaDict (bed + bath + location)
            key = line[2]+" "+line[3]+" "+line[9]
            if key in areaDict:
                areaDict[key][0] += area
                areaDict[key][1] += 1
            else:
                areaDict[key] = [area,1]
            #write information for areaDict (bed + bath)
            key = line[2]+" "+line[3]
            if key in areaDict:
                areaDict[key][0] += area
                areaDict[key][1] += 1
            else:
                areaDict[key] = [area,1]
            #write information for areaDict (bed + location)
            key = line[2]+" "+line[9]
            if key in areaDict:
                areaDict[key][0] += area
                areaDict[key][1] += 1
            else:
                areaDict[key] = [area,1]
        else:
            listobj.append(lineFive)
        listobj.append(line[6]) #Library
        listobj.append(line[7]) #Fitness Center
        listobj.append(line[8]) #Distance
        listobj.append(line[9]) #Location
        sheet_writer.writerow(listobj)
            
        
```

```python
#Fill in missing values for square feet
with open("datasetEdited.csv", 'r') as f, open('datasetFinal.csv','w') as writer:
    sheet_reader = csv.reader(f, delimiter=",")
    sheet_writer = csv.writer(writer, delimiter=",")
    sheet_writer.writerow(["ApartmentName","URL","Beds","Baths","Cost","SquareFeet",
                        "Library","Fitness Center","Distance","Location"])
    next(sheet_reader) #exclude header
    for line in sheet_reader:
        listobj = []
        for i in range(0,5):
            listobj.append(line[i])
        if line[5] == "":
            key1 = line[2]+" "+line[3]+" "+line[9]
            key2 = line[2]+" "+line[3]
            key3 = line[2]+" "+line[9]
            if key1 in areaDict:
                listobj.append(math.ceil(areaDict[key1][0] / areaDict[key1][1]))
            elif key2 in areaDict:
                listobj.append(math.ceil(areaDict[key2][0] / areaDict[key2][1]))
            else:
                listobj.append(math.ceil(areaDict[key3][0] / areaDict[key3][1]))
        else:
            listobj.append(line[5])
        for i in range(6,10):
            listobj.append(line[i])
        sheet_writer.writerow(listobj)
        
```
